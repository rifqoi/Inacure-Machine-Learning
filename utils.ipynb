```python
from sklearn.model_selection import train_test_split
folder = '/content/drive/Shareddrives/Bangkit Capstone/Final_Datasets/Ceri'
target_folder = f"/content/drive/Shareddrives/Bangkit Capstone/Final_Split_Datasets/"

files = os.listdir(folder)
files = [os.path.join(folder, file) for file in files]
train, val = train_test_split(files , train_size=0.7, test_size=0.3)
val, test = train_test_split(val, train_size=0.66, test_size=0.34)

for file in train:
    dest_dir = os.path.join(target_folder, 'train', os.path.basename(folder))
    dest_file = os.path.join(dest_dir, os.path.basename(file))
    if os.path.exists(dest_dir) is False:
        os.mkdir(dest_dir)

    print(f"copy {file} to {dest_file}")
    shutil.copy(file, dest_file)

for file in val:
    dest_dir = os.path.join(target_folder, 'validation', os.path.basename(folder))
    dest_file = os.path.join(dest_dir, os.path.basename(file))
    if os.path.exists(dest_dir) is False:
        os.mkdir(dest_dir)

    print(f"copy {file} to {dest_file}")
    shutil.copy(file, dest_file)


for file in test:
    dest_dir = os.path.join(target_folder, 'test', os.path.basename(folder))
    dest_file = os.path.join(dest_dir, os.path.basename(file))
    if os.path.exists(dest_dir) is False:
        os.mkdir(dest_dir)

    print(f"copy {file} to {dest_file}")
    shutil.copy(file, dest_file)
```

```python
dataset_dir = f"/content/drive/Shareddrives/Bangkit Capstone/Final_Split_Datasets"
train_dir = os.path.join(dataset_dir, "train")
validation_dir = os.path.join(dataset_dir, "validation")
test_dir = os.path.join(dataset_dir, "test")
```

# List Images

```python
def list_images():
  classes_dirs = os.listdir(train_dir)
  for folder in classes_dirs:
    train_classes = os.listdir(os.path.join(train_dir, folder))
    validation_classes = os.listdir(os.path.join(validation_dir, folder))
    test_classes = os.listdir(os.path.join(test_dir, folder))
    print(f"{folder}: \n \t{len(train_classes)} train images, {len(validation_classes)} validation images, and {len(test_classes)} test images,  total {len(train_classes) + len(validation_classes) + len(test_classes)} images")
    print()
  print(f"Total {len(classes_dirs)} class.")
list_images()
```
